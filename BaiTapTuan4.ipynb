```python
#W4A6
def prime(n):
  flag = True
  if n < 1:
    return False
  elif n == 1:
    return False
  elif n > 1:
    for i in range(2,n):
      if n % i == 0:
        flag = False
        break
    if flag == True:
      return True
    else:
      return False
a, b = map(int, input().split())
dem = 0
for i in range(a, b+1):
  if prime(i) == True:
    dem += 1
print(dem)
```

```python
#W4A7
n = int(input())
def prime(n):
  flag = True
  if n < 1:
    return False
  elif n == 1:
    return False
  elif n > 1:
    for i in range(2,n):
      if n % i == 0:
        flag = False
        break
    if flag == True:
      return True
    else:
      return False
lon = 0
for i in range(n):
  if prime(i) == True:
    lon = i
print(lon)
```

```python
#W4A8
n = int(input())
i = 0
def reverse(x):
  res = 0
  while x > 0:
    res = res * 10 + x % 10
    x = x // 10
  return res
while i <= 1000:
  n = n + reverse(n)
  i += 1
  if n == reverse(n):
    print(i, n)
    break
else:
  print("Số này không thể tạo thành palindrome")
```

```python
#W4A9
import math
n = int(input())
def chu_so(x):
  txt = str(x)
  a = list(txt)
  flag = True
  for i in range(len(a)):
    for j in range(i+1, len(a)):
      if a[i] == a[j]:
        flag = False
        break
  return flag
i = 1
while i < math.sqrt(n):
  if chu_so(i**2) == True:
    print(i**2, end = " ")
  i += 1
```

```python
#W4A10
n = int(input())
a = n
def collatz(x):
  i = 0
  while x != 1:
      if x % 2 == 0:
          x = x / 2
      else:
          x = 3*x + 1
      i += 1
  return i
for j in range(n-1, 2, -1):
  if collatz(j) > collatz(a):
      a = j
print(a, collatz(a))
```
